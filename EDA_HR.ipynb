**Creation de nouvelles colonnes**

il s'agit des colonnes **Âge_emp, Catégorie_âge**, et **tranche_salaire** et **niveau engagement**

```python
# calculer l'age des employés

date_now = datetime.now() # obtenir la date d'aujourd'hui

df['Age_emp'] = date_now.year - df['DOB'].dt.year

df['Age_emp'].head()
```

```python
# créer la fonction qui permet de donner la catégorie d'age dans laquelle chaque employé se situe en fonction de sa tranche d'age
def categorie_age(age):

    if 18 <= age <= 30:
        return "Jeune"
    elif 31 <= age <= 60 :
        return "Adulte"
    else :
        return "Vieux"
```

```python
# créer la colonne Catégorie_age

df['Catégorie_âge'] = df['Age_emp'].apply(categorie_age)

df[['Age_emp','Catégorie_âge']].head()
```


Le decoupage du salaire en tranche de la colonne **Tranche_salaire** que nous créerons se fera en fonction des **quartiles** de la colonne **Salary**.

```python
# Créer la colonne Tranche_Salaire

df['Tranche_Salaire'] = pd.qcut(df['Salary'], q=4, labels=['Bas', 'Moyen', 'Élevé', 'Très élevé'])

df[['Salary', 'Tranche_Salaire']].head()
```

```python
# Créer la colonne Tranche_Salaire

df['Niveau_engagement'] = pd.qcut(df['EngagementSurvey'], q=4, labels=['Bas', 'Moyen', 'Élevé', 'Très élevé'])

df[['EngagementSurvey', 'Niveau_engagement']].head()
```

```python
# faire un barplot pour visualiser l'écart salarial entre les hommes et les femmes

# Nettoyage de la colonne 'Sex' si nécessaire
df['Sex'] = df['Sex'].str.strip()

# Calculer la médiane des salaires par Département et Sexe
salaire_médian_département_sexe = df.groupby(['Department', 'Sex'])['Salary'].median().unstack(fill_value=0)

# Vérifier les colonnes obtenues
print(salaire_médian_département_sexe.columns)

# Vérification et correction des colonnes si nécessaire
if 'Male' in salaire_médian_département_sexe.columns:
    salaire_médian_département_sexe.rename(columns={'Male': 'M', 'Female': 'F'}, inplace=True)

# Exclure "Executive Office" et les départements où il manque un sexe
salaire_médian_département_sexe = salaire_médian_département_sexe.dropna()
salaire_médian_département_sexe = salaire_médian_département_sexe.drop(index='Executive Office', errors='ignore')

# Calculer l'écart salarial : (M - F)
salaire_médian_département_sexe['Ecart_H-F'] = salaire_médian_département_sexe['M'] - salaire_médian_département_sexe['F']

# Réinitialiser l'index pour faciliter le tracé
salaire_ecart = salaire_médian_département_sexe.reset_index()

# Déterminer les valeurs max et min pour équilibrer l'axe Y
ecart_max = salaire_ecart['Ecart_H-F'].max()
ecart_min = salaire_ecart['Ecart_H-F'].min()
lim = max(abs(ecart_max), abs(ecart_min))  # Valeur absolue max pour un affichage équilibré

# Palette divergente pour bien distinguer les écarts positifs et négatifs
palette = sns.diverging_palette(220, 20, n=len(salaire_ecart))

# Tracer le barplot de l'écart salarial
plt.figure(figsize=(12,6))
ax = sns.barplot(data=salaire_ecart, x='Department', y='Ecart_H-F', palette=palette)

# Ajouter des annotations sur les barres
for p in ax.patches:
    ax.annotate(
        f'{p.get_height():.0f}',  # Valeur arrondie
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='bottom' if p.get_height() > 0 else 'top',
        fontsize=10, color='black',
        xytext=(0, -10 if p.get_height() < 0 else 10),  # Décalage selon le signe
        textcoords='offset points'
    )

# Ligne de référence à 0
plt.axhline(0, color='black', linestyle='dashed')

# Ajustement des limites pour équilibrer l'affichage
plt.ylim(-lim, lim)

# Titres et labels
plt.title("Écart du salaire médian par sexe en fonction du département")
plt.xlabel("Département")
plt.ylabel("Écart salarial médian (M - F)")
plt.xticks(rotation=45)

# Afficher le graphique
plt.show()
```

Les **Hommes** ont un écart salarial médian supérieur à celui des **Femmes** dans les départements suivants : 
- Bureau exécutif avec **5520**
- Commercial avec **3978**
- IT/IS avec **614**.

Alors que les **Femmes** ont un écart salarial médian supérieur à celui des **Femmes** dans les départements suivants
- Software engineering avec **4481**
- Production avec **2014**.
